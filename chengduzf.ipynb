```python
import folium
from folium.plugins import HeatMap
import pandas as pd
import pymongo

# define the world map
world_map = folium.Map()

latitude = 30.67
longitude = 104.06

# Create map and display it
san_map = folium.Map(location=[latitude, longitude], zoom_start=11)

client = pymongo.MongoClient('mongodb://localhost/dev')
db = client['dev']
collection = db['cdzf']

heatData = []
for i in collection.find():
    if float(i['price']) > 900 and float(i['price']) < 1500:
        folium.Marker(
            location=[i['curLatitude'], i['curLongitude']],
            icon=folium.Icon(color='red')
        )
    heatData.append([i['curLatitude'], i['curLongitude'], float(i['price'])])
    #folium.Circle(
    #    radius=10,
    #    location=[i['curLatitude'], i['curLongitude']]
    #).add_to(san_map)
#print(heatData)
HeatMap(heatData, radius=6).add_to(san_map)

folium.Marker(
    location=[30.554107, 104.077616],
    popup='天府软件园',
    icon=folium.Icon(color='blue')
).add_to(san_map)

folium.Marker(
    location=[30.774647, 104.107501],
    popup='中铁产业园',
    icon=folium.Icon(color='black')
).add_to(san_map)


# Display the map of San Francisco
san_map
```
